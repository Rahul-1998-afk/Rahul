## 4) Remove rows which have negative values in Quantity column

## 5) Add the columns - Month, Day and Hour for the invoice

```python
ecom_data["InvoiceDate"] = pd.to_datetime(ecom_data["InvoiceDate"])
#Changing from object to datetime dtype
```

```python
ecom_data["date"] = ecom_data["InvoiceDate"].apply(lambda x: str(x).split(" "))
#Or ecom_date["date"]=pd.DatetimeIndex(df['InvoiceDate']).date

#Splitting date from InvoiceDate
```

```python
ecom_data["year"]=pd.DatetimeIndex(ecom_data['InvoiceDate']).year
ecom_data["month"]=pd.DatetimeIndex(ecom_data['InvoiceDate']).month
ecom_data["day"]=pd.DatetimeIndex(ecom_data['InvoiceDate']).day
```
